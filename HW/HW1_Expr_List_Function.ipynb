กำหนดข้อมูลของกรอบสี่เหลี่ยมเก็บจำนวนจริง 4 ตัวในรูปแบบ $[x, y, w, h]$ เมื่อ
- $x$, $y$ เป็นพิกัดมุมบนซ้าย 
- $w$ เป็นความกว้างและ

ให้นิสิตเขียนฟังก์ชัน 5 ฟังก์ชัน ดังนี้
1. `center(r)` คืนพิกัด `(x, y)` ของจุดตัดของเส้นทแยงมุมสองเส้นของกรอบสี่เหลี่ยม เป็นลิสต์ `[x, y]` 
2. `distance(r1, r2)` คืนระยะห่างของจุดศูนย์กลางของกรอบสี่เหลี่ยม `r1` และ `r2` 
3. `intersection(r1, r2)` คืนค่าพื้นที่ที่ `r1` กับ `r2` ซ้อนกัน (ให้คิดโดยกำหนดว่า `r1` กับ `r2` ซ้อนกัน แต่หากสามารถคำนวณในกรณีที่ไม่ซ้อนกันได้ จะได้โบนัส 10%)
4. `union(r1, r2)` คืนค่าพื้นที่ที่ครอบคลุมโดย `r1` หรือ `r2` (ในกรณีโบนัส 10% ที่ `r1` และ `r2` ไม่ซ้อนกัน จะเท่ากับ ผลรวมของพื้นที่ของ `r1`, `r2`)
5. `iou(r1, r2)` คืนค่าอัตราส่วนพื้นที่ที่ `r1` กับ `r2` ซ้อนกันต่อพื้นที่ที่ครอบคลุมโดย `r1` หรือ `r2`

````python
# HW01 (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

def center(r):
    x, y, w, h = r
    return [x+w/2, y-h/2]



#======================================
def distance(r1, r2):
    r1Center = center(r1)
    r2Center = center(r2)
    return ((r1Center[0]-r2Center[0])**2 + (r1Center[1]-r2Center[1])**2)**(0.5)








#======================================
def intersection(r1, r2):
    w1 = max(r1[0], r2[0])
    w2 = min(r1[0] + r1[2], r2[0] + r2[2])
    w = max(0, w2 - w1)
    l1 = min(r1[1], r2[1])
    l2 = max(r1[1] - r1[3], r2[1] - r2[3])
    l = max(0, l1 - l2)
    itsc = w*l
    return itsc







#======================================
def union(r1, r2):
    return r1[2]*r1[3] + r2[2]*r2[3] - intersection(r1, r2);







#======================================
def iou(r1, r2):
    return intersection(r1, r2)/union(r1, r2);







#======================================
````
