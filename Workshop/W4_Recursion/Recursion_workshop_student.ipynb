```
from IPython.display import Image
from graphviz import Digraph
```

```
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None       
```

```
def visualize_tree(tree):
    def add_nodes_edges(tree, dot=None):
        # สร้าง Digraph object
        if dot is None:
            dot = Digraph()
            dot.node(name=str(tree), label=str(tree.val))

        # เพิ่ม nodes
        if tree.left:
            dot.node(name=str(tree.left) ,label=str(tree.left.val))
            dot.edge(str(tree), str(tree.left))
            dot = add_nodes_edges(tree.left, dot=dot)
            
        if tree.right:
            dot.node(name=str(tree.right) ,label=str(tree.right.val))
            dot.edge(str(tree), str(tree.right))
            dot = add_nodes_edges(tree.right, dot=dot)

        return dot
    
    # เพิ่ม nodes วนไปเรื่อยๆ และสร้าง list ของ edges
    dot = add_nodes_edges(tree)

    # Visualize the graph
    display(dot)
```

**ตัวอย่างการใช้ TreeNode และฟังก์ชั่น visualize_tree**

```
t1 = TreeNode(1)
t1.left = TreeNode(3)
t1.left.left = TreeNode(5)
t1.right = TreeNode(2)
```

```
visualize_tree(t1)
```

ในส่วนนี้ เราต้องการที่จะสร้างฟังก์ชั่นที่ใช้เช็คความแตกต่างของประโยค 2 ประโยค

---


เราต้องสร้าง **Minimum Edit Distance function** เพื่อคำนวณหาค่าความแตกต่างระหว่างข้อความ 2 อัน



\\
ในส่วนของการทำงานของ Minimum Edit Distance นั้น, สมมุติว่าเรามีสตริง 2 สตริง \\
**m** คือ ความยาวของสตริงที่ 1 \\
**n** คือ ความยาวของสตริงที่ 2

เราจะสนใจทุกๆการเปรียบเทียบบนตัวอักษรตัวสุดท้ายของสตริงที่ 1 โดยเราจะคำนวณวนไปเรื่อยๆเพื่อ**หาค่าที่น้อยที่สุดของการเปรียบเทียบทั้ง 3 แบบ** \\

โดยการเปรียบเทียบทั้ง 3 แบบได้แก่

* **การเติม: เราจะให้ฟังก์ชั่นรับ m และ n-1** 
* **การลบ: เราจะให้ฟังก์ชั่นรับ m-1 และ n**
* **การแทนที่: เราจะให้ฟังก์ชั่นรับ m-1 และ n-1**

ในกรณี **สตริงเปล่าเทียบกับสตริงอื่นให้คืนความยาวของสตริงที่ไม่ใช่สตริงเปล่า** กลับมา

**ย้ำาา: ทุกๆการเปรียบเทียบจะมีค่าการทำงานเท่ากับ 1 !!**
