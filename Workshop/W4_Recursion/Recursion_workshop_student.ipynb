```python
from IPython.display import Image
from graphviz import Digraph
import os
os.environ["PATH"] += os.pathsep + 'C:\Program Files\Graphviz\bin'
```

```python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None       
```

```python
def visualize_tree(tree):
    def add_nodes_edges(tree, dot=None):
        # สร้าง Digraph object
        if dot is None:
            dot = Digraph()
            dot.node(name=str(tree), label=str(tree.val))

        # เพิ่ม nodes
        if tree.left:
            dot.node(name=str(tree.left) ,label=str(tree.left.val))
            dot.edge(str(tree), str(tree.left))
            dot = add_nodes_edges(tree.left, dot=dot)
            
        if tree.right:
            dot.node(name=str(tree.right) ,label=str(tree.right.val))
            dot.edge(str(tree), str(tree.right))
            dot = add_nodes_edges(tree.right, dot=dot)

        return dot
    
    # เพิ่ม nodes วนไปเรื่อยๆ และสร้าง list ของ edges
    dot = add_nodes_edges(tree)

    # Visualize the graph
    display(dot)
```

**ตัวอย่างการใช้ TreeNode และฟังก์ชั่น visualize_tree**

```python
t1 = TreeNode(1)
t1.left = TreeNode(3)
t1.left.left = TreeNode(5)
t1.right = TreeNode(2)
```

```python
visualize_tree(t1)
```

### 1.1 สร้างฟังก์ชั่นที่ทำงานตามสมการฟังก์ชั่นด้านล่างโดยใช้เพียงแค่ for loop###

**ตัวอย่าง:** \\
* factorial(3) --> 6 \\
* factorial(4) --> 24 \\
* factorial(10) --> 3628800

```python
# TODO 1.1 สร้างฟังก์ชั่น
def factorial(value):
  
  ## เติมโค้ดด้านล่าง ##
  mul_value = 1
  for i in range(1, value):
    mul_value *= i 
  
  ##############
  
  return mul_value   
```

```python
print(factorial(3))
print(factorial(4))
print(factorial(10))
```

### 1.2 สร้างฟังก์ชั่น recursive ที่ทำงานตามสมการฟังก์ชั่นด้านล่าง ### 


![alt text](https://i.imgur.com/8iFS5Hu.png)



**ตัวอย่าง:** \\
* factorial(3) --> 6 \\
* factorial(4) --> 24 \\
* factorial(10) --> 3628800

```python
# TODO 1.2.1 สร้างฟังก์ชั่น recursive
def factorial(value):
  
  if value == 1:
    return value
  
  ## เติมโค้ดด้านล่าง ## 
  
  mul_value = factorial(value - 1) * value
  
  ##############
  
  return mul_value  
```

```python
print(factorial(3))
print(factorial(4))
print(factorial(10))
```

```python
# TODO 1.2.2 สร้างฟังก์ชั่น recursive โดยที่ให้สร้าง tree ของฟังก์ชั่นด้วย
def factorial(value):
  
  ## เติมโค้ดด้านล่าง ##


  if value == 1:
    return value, TreeNode(value)
  
  cFac, cTree = factorial(value - 1)
  mul_value = value * cFac
  t = TreeNode(value)
  t.right = cTree 

  ##############
  
  return mul_value,t  
```

```python
fac,tree = factorial(3)
visualize_tree(tree)
print(fac)
```

### 1.3 สร้างฟังก์ชั่น recursive ที่ทำงานตามสมการฟังก์ชั่นด้านล่าง ### 

![alt text](https://i.imgur.com/uVpuc32.png)

**ตัวอย่าง:** \\
* fibonacci(4) --> 3 
* fibonacci(10) --> 55 
* fibonacci(30) --> 832040 


```python
# TODO 1.3.1 สร้างฟังก์ชั่น recursive
def fibonacci(value):
  
  ## เติมโค้ดด้านล่าง ##
  if value == 0:
    return 0
  elif value == 1:
    return 1

  sum_value = fibonacci(value - 1) + fibonacci(value - 2)
  ##############
  
  return sum_value  
```

```python
print(fibonacci(4))
print(fibonacci(10))
print(fibonacci(30))
```

```python
# TODO 1.3.2 สร้างฟังก์ชั่น recursive โดยที่ให้สร้าง tree ของฟังก์ชั่นด้วย

def fibonacci(value):
  
  ## เติมโค้ดด้านล่าง ##
  if value == 0:
    return 0, TreeNode(0)
  elif value == 1:
    return 1, TreeNode(1)

  tmp1_value, tmp1Tree = fibonacci(value - 1)
  tmp2_value, tmp2Tree = fibonacci(value - 2)

  sum_value = tmp1_value + tmp2_value
  t = TreeNode(sum_value)
  t.left = tmp1Tree
  t.right = tmp2Tree
  
  ##############
  
  return sum_value,t  
```

```python
fibo,tree = fibonacci(5)
visualize_tree(tree)
print(fibo)
```

# Edit Distance

ในส่วนนี้ เราต้องการที่จะสร้างฟังก์ชั่นที่ใช้เช็คความแตกต่างของประโยค 2 ประโยค

---


เราต้องสร้าง **Minimum Edit Distance function** เพื่อคำนวณหาค่าความแตกต่างระหว่างข้อความ 2 อัน



\\
ในส่วนของการทำงานของ Minimum Edit Distance นั้น, สมมุติว่าเรามีสตริง 2 สตริง \\
**m** คือ ความยาวของสตริงที่ 1 \\
**n** คือ ความยาวของสตริงที่ 2

เราจะสนใจทุกๆการเปรียบเทียบบนตัวอักษรตัวสุดท้ายของสตริงที่ 1 โดยเราจะคำนวณวนไปเรื่อยๆเพื่อ**หาค่าที่น้อยที่สุดของการเปรียบเทียบทั้ง 3 แบบ** \\

โดยการเปรียบเทียบทั้ง 3 แบบได้แก่

* **การเติม: เราจะให้ฟังก์ชั่นรับ m และ n-1** 
* **การลบ: เราจะให้ฟังก์ชั่นรับ m-1 และ n**
* **การแทนที่: เราจะให้ฟังก์ชั่นรับ m-1 และ n-1**

ในกรณี **สตริงเปล่าเทียบกับสตริงอื่นให้คืนความยาวของสตริงที่ไม่ใช่สตริงเปล่า** กลับมา

**ย้ำาา: ทุกๆการเปรียบเทียบจะมีค่าการทำงานเท่ากับ 1 !!**

### 2.สร้างฟังก์ชั่น Minimum Edit Distance function ### 

```python
# TODO 2.1 สร้างฟังก์ชั่น recursive ของ Minimum Edit Distance
def edit_distance(real,fake,m,n):
  
  if m == 0:
    return n #เติมสิ่งที่ฟังก์ชั่นต้องคืนกลับมา เมื่อสตริงตัวที่ 1 ที่ให้เช็คเป็นสตริงเปล่า
  
  if n == 0: 
    return m #เติมสิ่งที่ฟังก์ชั่นต้องคืนกลับมา เมื่อสตริงตัวที่ 2 ที่ให้เช็คเป็นสตริงเปล่า
  
  #เติมสิ่งที่ฟังก์ชั่นต้องคืนกลับมา ในกรณีที่ตัวสุดท้ายของสตริงทั้ง 2 อันเหมือนกันให้ข้ามไปคำนวณตัวถัดไป
  if real[m-1] == fake[n-1]: 
    return edit_distance(real[:m-1], fake[:n-1], m-1, n-1)
  
  #เติมสิ่งที่ฟังก์ชั่นต้องคืนค่าให้สมบูรณ์ เพื่อที่ฟังก์ชั่นจะได้ทำงานถูกต้อง
  return min(
            edit_distance(real, fake[:n-1], m, n-1)
            ,edit_distance(real[:m-1], fake, m-1, n)
            ,edit_distance(real[:m-1], fake[:n-1], m-1, n-1)
  ) + 1
```

```python
#ในการทดสอบฟังก์ชั่น เราจะทดสอบบนสตริงเล็กๆ เช่น CAT กับ DOG
print(edit_distance("CAT","DOG",3,3))
print(edit_distance("COMP","CHAMP",4,5))
print(edit_distance("PRAYUT","PAYDAY",6,6))
```

```python
#เนื่องจาก Minimum Edit Distance ใช้เวลานานมากหากเปรียบเทียบสตริงยาวๆ
#เราจึงแบ่งประโยคออกเป็นส่วนๆแทน

real_meme = ["makes a wish on a falling star","smash by a meteor"] 

fake_meme = ["bakes a witch on a falling star","smash bye a mateor"]
```

```python
## TODO 3 ลองเขียนในรูปแบบของ For loop ดู

def edit_ds_dp(real,fake):
    n = len(real)
    m = len(fake)

    prev = [ 0 for c in range(m+1) ]
    for r, realChar in enumerate(real, start = 1):
        cur = [ 0 for c in range(m+1) ]
        cur[0] = r
        for c, fakeChar in enumerate(fake, start = 1):
            if realChar == fakeChar:
                cur[c] = prev[c-1]
            else:
                cur[c] = min(
                    cur[c-1]
                    ,prev[c]
                    ,prev[c-1]
                ) + 1

        prev = cur
    return prev[m]

edit_ds_dp(real_meme[0], fake_meme[0])
```

```python
for i in range(len(real_meme)):
  print(real_meme[i]+' VS '+fake_meme[i])
  print('Edit Distance: '+str(edit_ds_dp(real_meme[i],fake_meme[i])))
```
