# TO DO 5: หาค่ามัธยฐานและค่าเฉลี่ยของแต่ละวิชา และนำไปแสดงผลบนกราฟแท่งแบบ Multiple plot

```python
#Write code below please.
mean_math = math_score.mean()
mean_reading = reading_score.mean()
mean_writing = writing_score.mean()

median_math = math_score.median()
median_reading = reading_score.median()
median_writing = writing_score.median()
```

```python
#Plot multiple bar graph
#Write code below please.
import numpy as np

plt.figure(figsize=(14,8))
X = ["Math", "Reading", "Writing"]
X_axis = np.arange(len(X))

width = 0.4
mean_scores = [mean_math, mean_reading, mean_writing]
median_scores = [median_math, median_reading, median_writing]
plt.bar(X_axis, mean_scores, width)
plt.bar(X_axis + width, median_scores, width)

plt.legend(["mean score", "median score"])
plt.title('Mean and Median Score')
plt.ylabel('Scores')
plt.xticks(X_axis + width/2, X)
```
